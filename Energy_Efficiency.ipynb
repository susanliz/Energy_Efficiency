```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.multioutput import RegressorChain
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import RidgeCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

from sklearn.model_selection import cross_val_score
from numpy import absolute, mean, std
from sklearn.model_selection import RepeatedKFold
import pickle
```

```python
df = pd.read_excel("ENB2012_data.xlsx")
df.columns = ["Relative_Compactness","Surface_Area","Wall_Area",
                "Roof_Area",
                "Overall_Height",
                "Orientation",
                "Glazing_Area",
                "Glazing_Area_Distribution",
                "Heating_Load",
                "Cooling_Load"]
```

```python
testdata=[[0.62,	808.5,	367.5,	220.50,	3.5,	5,	0.4,	5]]
```

```python
class model:
  def models(self):
    print('Linear Regression :')
    X = df.drop(['Heating_Load','Cooling_Load'],axis=1)
    y = df[['Heating_Load','Cooling_Load']]
    X = X.values 
    
    lr = LinearRegression()
    lr.fit(X,y)
    lr.predict(testdata)
    lrp = lr.predict(testdata)
    print('Predicted value :',lrp)

    cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
    n_scores_lr = cross_val_score(lr, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
    n_scores_lr = absolute(n_scores_lr)
    print('MAE: %.3f (%.3f)' % (mean(n_scores_lr), std(n_scores_lr)),'\n')

    print('---------------------------------------------------------------------------','\n')

    print('KNeighborsRegressor :')
    KN = KNeighborsRegressor()
    KN.fit(X, y)
    KNP = KN.predict(testdata)
    print('Predicted value :',KNP)
    
    n_scores_KN = cross_val_score(KN, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
    n_scores_KN = absolute(n_scores_KN)
    print('MAE: %.3f (%.3f)' % (mean(n_scores_KN), std(n_scores_KN)),'\n')

    print('---------------------------------------------------------------------------','\n')

    print('ElasticNet :')
    E = ElasticNet()
    wrapperE = RegressorChain(E)
    wrapperE.fit(X, y)
    EP = wrapperE.predict(testdata)
    print('Predicted value :',EP)

    n_scores_E= cross_val_score(wrapperE, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
    n_scores_E = absolute(n_scores_E)
    print('MAE: %.3f (%.3f)' % (mean(n_scores_E), std(n_scores_E)),'\n')

    print('---------------------------------------------------------------------------','\n')

    print('RidgeCV :')
    CV = RidgeCV()
    CV = RegressorChain(CV)
    CV.fit(X, y)
    CVP = CV.predict(testdata)
    print('Predicted value :',CVP)

    n_scores_CV = cross_val_score(CV, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
    n_scores_CV = absolute(n_scores_CV)
    print('MAE: %.3f (%.3f)' % (mean(n_scores_CV), std(n_scores_CV)),'\n')

    print('---------------------------------------------------------------------------','\n')

    print('DecisionTreeRegressor :')
    DTR = DecisionTreeRegressor()
    DTR.fit(X, y)
    DTRP = DTR.predict(testdata)
    print('Predicted value :',DTRP)

    n_scores_DTR = cross_val_score(DTR, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
    n_scores_DTR = absolute(n_scores_DTR)
    print('MAE: %.3f (%.3f)' % (mean(n_scores_DTR), std(n_scores_DTR)),'\n')
    pickle.dump(DTR,open('mode_DecisionTreeRegressor.pkl','wb')) 

    print('---------------------------------------------------------------------------','\n')

    print('RandomForestRegressor :')
    RF = RandomForestRegressor()
    RF.fit(X, y)
    RFP = RF.predict(testdata)
    print('Predicted value :',RFP)

    n_scores_RF = cross_val_score(RF, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
    n_scores_RF = absolute(n_scores_RF)
    print('MAE: %.3f (%.3f)' % (mean(n_scores_RF), std(n_scores_RF)),'\n')
    print('---------------------------------------------------------------------------','\n')





obj_model = model()
```
